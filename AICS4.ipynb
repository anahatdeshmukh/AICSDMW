```python
# Assignment 4
# Perform Feature Engineering Operations on Raw Data

# Import libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

# ---------------- LOAD DATASET ----------------

# Load CSV dataset
df = pd.read_csv("employee_dataset.csv")

# Display original dataset
print("Original Dataset:\n")
print(df.head())

# ---------------- HANDLE MISSING VALUES ----------------

# Fill missing numerical values with mean
df['Age'] = df['Age'].fillna(df['Age'].mean())

df['Salary'] = df['Salary'].fillna(df['Salary'].mean())

# ---------------- LABEL ENCODING ----------------

# Convert categorical text data into numbers
encoder = LabelEncoder()

df['Department'] = encoder.fit_transform(df['Department'])

# ---------------- FEATURE SCALING ----------------

# Normalize numerical columns
scaler = StandardScaler()

df[['Age', 'Salary']] = scaler.fit_transform(
    df[['Age', 'Salary']]
)

# ---------------- DISPLAY FINAL DATA ----------------

print("\nProcessed Dataset:\n")
print(df.head())

# ---------------- SAVE PROCESSED DATA ----------------

df.to_csv("processed_employee_dataset.csv", index=False)

print("\nProcessed dataset saved successfully.")
```
