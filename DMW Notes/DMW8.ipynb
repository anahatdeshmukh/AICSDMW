```python
from itertools import combinations

# Dataset
transactions = [
    ['Milk', 'Bread', 'Butter'],
    ['Bread', 'Butter'],
    ['Milk', 'Bread'],
    ['Milk', 'Butter'],
    ['Bread']
]

min_support = 2
min_confidence = 0.5

# -----------------------------
# Find unique items
# -----------------------------
items = set()

for transaction in transactions:
    for item in transaction:
        items.add(item)

# -----------------------------
# Frequent 1-itemsets
# -----------------------------
freq1 = {}

for item in items:
    count = 0

    for transaction in transactions:
        if item in transaction:
            count += 1

    if count >= min_support:
        freq1[(item,)] = count

print("Frequent 1-itemsets:")
print(freq1)

# -----------------------------
# Frequent 2-itemsets
# -----------------------------
freq2 = {}

pairs = combinations(items, 2)

for pair in pairs:
    count = 0

    for transaction in transactions:
        if set(pair).issubset(set(transaction)):
            count += 1

    if count >= min_support:
        freq2[pair] = count

print("\nFrequent 2-itemsets:")
print(freq2)

# -----------------------------
# Association Rules
# -----------------------------
print("\nAssociation Rules:")

for pair in freq2:
    a, b = pair

    support_ab = freq2[pair]

    support_a = freq1[(a,)]
    support_b = freq1[(b,)]

    confidence_a_b = support_ab / support_a
    confidence_b_a = support_ab / support_b

    if confidence_a_b >= min_confidence:
        print(f"{a} -> {b}  Confidence = {confidence_a_b:.2f}")

    if confidence_b_a >= min_confidence:
        print(f"{b} -> {a}  Confidence = {confidence_b_a:.2f}")
```
